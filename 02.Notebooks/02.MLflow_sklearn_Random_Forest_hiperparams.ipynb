```python
import mlflow
import mlflow.sklearn

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import * 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
creditos = pd.read_csv('../01.Dados/Credit.csv')
creditos
```

```python
creditos.head()
```

```python
for col in creditos.columns:
    if creditos[col].dtype == 'object':
        creditos[col] = creditos[col].astype('category').cat.codes
creditos.head()
```

```python
previsores = creditos.iloc[:,0:20].values
classe = creditos.iloc[:,20].values
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe, test_size=0.2, random_state=123)
```

```python
def tain_rf(n_estimators):
    
    # Qual o nome do conjunto de experimentos.
    mlflow.set_experiment("rfexperimento")
    # Clásula with para definir o contexto do modelo.
    # Agora dessa vez, vamos guardar os hiperparâmetros.
    with mlflow.start_run():

        modelRF = RandomForestClassifier(n_estimators = n_estimators)
        modelRF.fit(X_treinamento,y_treinamento)
        previsoes = modelRF.predict(X_teste)

        #log de hiper parametros
        #mlflow.log_param("n_estimators",n_estimators)
        list(map(lambda x: mlflow.log_param((x[0]), x[1]), modelRF.get_params().items()))

        #métricas
        f1 = f1_score(y_teste, previsoes)
        log = log_loss(y_teste, previsoes)
        auc = roc_auc_score(y_teste, previsoes)    
        recall = recall_score(y_teste, previsoes)
        acuracia = accuracy_score(y_teste, previsoes)
        precision = precision_score(y_teste, previsoes)    

        # Registrar métricas
        mlflow.log_metric('f1', f1)
        mlflow.log_metric('auc', auc)
        mlflow.log_metric('log', log)
        mlflow.log_metric('recall', recall)
        mlflow.log_metric('acuracia', acuracia)
        mlflow.log_metric('precision', precision)

        #gráficos
        #*Dica: é importante colocar o nome das imagens diferentes para não sobrescrever com outros gráficos.
        roc = plot_roc_curve(modelRF, X_teste, y_teste)    
        plt.savefig('rocRF.png')
        confusion = plot_confusion_matrix(modelRF, X_teste, y_teste)    
        plt.savefig('confusionRF.png')

        #logar gráficos
        mlflow.log_artifact('rocRF.png')    
        mlflow.log_artifact('confusionRF.png')

        #modelo
        mlflow.sklearn.log_model(modelRF, "modelRF")

        #informaçoes da execução. Cada model executado tem um identificador unico.
        print('Modelo: ',mlflow.active_run().info.run_uuid)

    mlflow.end_run()
```

```python
arvores = [50,100,500,750,1000]
for n in arvores:
    tain_rf(n_estimators = n)
```
